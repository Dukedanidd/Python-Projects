```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame()
df['Edad'] = [25, 30, 35, 40, 45] 
df['Salario'] = [3000, 3500, 4000, 4500, 5000]

plt.title('Relación entre Edad y Salario')
plt.scatter(df['Edad'], df['Salario'], color='blue')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True)
            
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
# Crear DataFrame
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]}
df = pd.DataFrame(data)
```

```python
plt.title('Relación entre Edad y Salario')
plt.scatter(df['Edad'], df['Salario'])
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True)
plt.show
```

```python
X = df[['Edad']]
y = df['Salario']

modelo = linear_model.LinearRegression()
modelo.fit(X, y)
```

```python
plt.title('Regresión Lineal Simple')
plt.plot(df['Edad'], modelo.predict(X))
```

```python
# Combinar ambos gráficos
plt.scatter(df['Edad'], df['Salario'])
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True)
plt.plot(df['Edad'], modelo.predict(X))
```

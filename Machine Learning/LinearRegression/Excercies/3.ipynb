```python
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
data = {'Edad': [25, 30, 35, 40, 45],
        'Salario': [3000, 3500, 4000, 4500, 5000]}
df = pd.DataFrame(data)
```

```python
X  = df[['Edad']]
y = df['Salario']

modelo = linear_model.LinearRegression()
modelo.fit(X, y)
```

```python
plt.title('Regresion Lineal SSimple')
plt.scatter(df['Edad'], df['Salario'], color='blue')
plt.xlabel('Edad')
plt.ylabel('Salario')
plt.grid(True)
plt.plot(df['Edad'], modelo.predict(X))
```

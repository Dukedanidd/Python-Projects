```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Aqui estoy  cargando el dataset que voy a estar utilizando y visualizo los primeros valores
diamantes = sns.load_dataset('diamonds')
diamantes.head()
```

```python
# Definir el estilo de todos los graficos
sns.set(style="whitegrid")

# Creando el grafico relacional
g = sns.relplot(
    data = diamantes,
    x = "carat",
    y = 'price',
    kind = 'scatter',
    height = 6,
    aspect = 1.5
)

# Poniendole un titulo
g.fig.suptitle('Relacion entre el precio y el peso de los diamantes',
               va = 'baseline',
               ha = 'center')

g.set_axis_labels('Peso en Quilates', 'Precio en dolls')
```
